```python
import pandas as pd
import re
```

```python
df = pd.read_excel(r"/home/user/Documents/HOC/PYTHON/Data/DataCleaning/source/Customer Call List.xlsx")
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns='Not_Useful_Column')
```

```python
df["Last_Name"]=df["Last_Name"].str.strip('_/.')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-'  + x[6:10])
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
```

```python
df[['Street_Address','State','Zip_Code']] = df['Address'].str.split(',',expand=True)
```

```python
df = df.drop(columns='Address')
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace(r"^Y.*","Y", regex=True)
df['Paying Customer'] = df['Paying Customer'].str.replace(r"^N.*","N", regex=True)
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace(r"^Y.*","Y", regex=True)
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace(r"^N.*","N", regex=True)
```

```python
df = df.fillna('')
```

```python
df['Do_Not_Contact'].replace('', 'Y', inplace=True)
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y' or df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```
